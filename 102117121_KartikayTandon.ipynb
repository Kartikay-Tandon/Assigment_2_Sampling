# Step - 4 Applying five different sampling techniques

```python
import numpy as np
```

## Simple Random Sampling

```python
population_size = 1526
```

```python
sample_size = 1.96*1.96*0.5*0.5/(0.05*0.05)
```

```python
sample_size = int(sample_size)
```

```python
sample_size
```

```python
simple_random_sample = np.random.choice(population_size, size=sample_size, replace=False)
```

```python
simple_random_sample
```

```python
simple_random_sample_df = my_data_set_balanced.iloc[simple_random_sample].copy()
```

```python
simple_random_sample_df
```

## Stratified Sampling

```python
stratified_sample_size = 1.96*1.96*0.5*0.5*1*1/(0.05*0.05)
```

```python
stratified_sample_size = int(stratified_sample_size)
```

```python
stratified_sample_size
```

```python
# Define strata proportions and sizes
strata_proportions = [0.2, 0.3, 0.5,1]  # Adjust based on your data
strata_sizes = [int(prop * len(my_data_set_balanced)) for prop in strata_proportions]
# Calculate sample size for each stratum
total_sample_size = 384
stratum_sample_sizes = [int(prop * total_sample_size) for prop in strata_proportions]
# Initialize an empty DataFrame for the stratified sample
stratified_sample_df = pd.DataFrame()
# Stratified sampling
for stratum_value, stratum_size, stratum_sample_size in zip(my_data_set_balanced['Class'].unique(), strata_sizes, stratum_sample_sizes):
    stratum_data = my_data_set_balanced[my_data_set_balanced['Class'] == stratum_value]  # Adjust based on your data
    stratum_sample_index = np.random.choice(stratum_data.index, size=stratum_sample_size, replace=False)
    stratified_sample_df = pd.concat([stratified_sample_df, my_data_set_balanced.loc[stratum_sample_index]])
```

```python
stratified_sample_df
```

## Cluster Sampling

```python
# Define the number of clusters and calculate cluster size
num_clusters = 27 # Adjust based on your data
cluster_size = len(my_data_set_balanced) // num_clusters
# Calculate the sample size for each cluster
total_sample_size = 952
cluster_sample_size = total_sample_size // num_clusters
# Initialize an empty DataFrame for the cluster sample
cluster_sample_df = pd.DataFrame()
# Cluster sampling
selected_clusters = np.random.choice(range(num_clusters), size=cluster_sample_size, replace=True)
for cluster_index in selected_clusters:
    cluster_start = cluster_index * cluster_size
    cluster_end = (cluster_index + 1) * cluster_size
    cluster_data = my_data_set_balanced.iloc[cluster_start:cluster_end, :]
    cluster_sample_df = pd.concat([cluster_sample_df, cluster_data])
```

```python
cluster_sample_df
```

## Bootstrap sampling

```python
num_bootstrap_samples = 1
bootstrapped_sample_df = pd.DataFrame()
for _ in range(num_bootstrap_samples):
    bootstrap_sample = my_data_set_balanced.sample(n=len(my_data_set_balanced), replace=True)
    bootstrapped_sample_df = pd.concat([bootstrapped_sample_df, bootstrap_sample])
```

```python
bootstrapped_sample_df
```

## K-fold cross validation sampling

```python
from sklearn.model_selection import KFold
```

```python
num_folds = 2
# Initialize an empty DataFrame for the cross-validation sample
cross_validation_sample_df = pd.DataFrame()
# Cross-validation sampling
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)
for train_index, test_index in kf.split(my_data_set_balanced):
    fold_train_data = my_data_set_balanced.iloc[train_index, :]
    fold_test_data = my_data_set_balanced.iloc[test_index, :]
    cross_validation_sample_df = pd.concat([cross_validation_sample_df, fold_test_data])
```

```python
cross_validation_sample_df
```
